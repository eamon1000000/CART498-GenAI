```python
#API key setup
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))

numWrongs = 0
temp = 0.5

def selfDep_calculator():
    global numWrongs, temp

    print("="*40)
    print("=== Calculator of self-dep路re路ca路tion (deprkSHn) ===")
    print("="*40)
    while True:
        print("Enter an integer value 'n' for me to be tested on")
        print("n = ")
        user_n = input(">").strip()
        if user_n.isdigit():
            break
        else:
            print("Invalid input. Please enter an integer.")
    while True:
        print("Enter an integer value 'i' for me to be tested on")
        print("i = ")
        user_i = input(">").strip()
        if user_i.isdigit():
            break
        else:
            print("Invalid input. Please enter an integer.")

    print("Great, so n=" + user_n + " and i=" + user_i + " now...allow me to operate\n")

    string_to_pass = "n ="+user_n+"\ni ="+user_i


    response = client.responses.create(
        model="gpt-4.1-nano",
        temperature=0.0,
        top_p=0.0,
        input=[
            {
                "role": "developer",
                "content": """You are a specialized calculator that performs iterative self-multiplication.

                            TASK:
                            Given two integers (n, i):
                            - n = base number
                            - i = number of iterations

                            OPERATION:
                            Perform i iterations of squaring:
                            1. Start with n
                            2. Square the result (multiply by itself)
                            3. Repeat step 2 for i total iterations

                            EXAMPLE:
                            Input: n=2, i=3
                            Iteration 1: 2 * 2 = 4
                            Iteration 2: 4 * 4 = 16
                            Iteration 3: 16 * 16 = 256
                            Output: 256

                            OUTPUT FORMAT:
                            - Show each iteration step on a new line
                            - Final line shows only the final answer
                            - Format: "Step 1: [calculation] = [result]"
                            - Last line: "My final answer: [answer]"
                            - Output the number as a whole number, no conversion to decimal with eX notation

                            RULES:
                            - Only accept two integers
                            - If input is invalid, respond with 'ERROR'
                            - No explanations beyond the calculation steps"""
            },
            {
                "role": "assistant",
                "content": ""
            },
            {
                "role": "user",
                "content": string_to_pass
            }
        ]
    )
    full_response = response.output[0].content[0].text
    print(full_response)

    final_answer = full_response.strip().split()[-1]

    print("\nNow I will compare my final answer "+ final_answer+", with the actual final answer from my trusty REAL calculator")

    actual_result = int(user_n)
    print(f"\nActual Calculation:")

    for iteration in range(int(user_i)):
        previous = actual_result
        actual_result = actual_result * actual_result
        print(f"Step {iteration + 1}: {previous} * {previous} = {actual_result}")

    print(f"Actual Final Result: {actual_result}")

    new_prompt = ""

    if int(final_answer) == actual_result:
        new_prompt = "You got the answer correct, celebrate accordingly"
    else:
        numWrongs = numWrongs + 1
        temp = 0.5 * numWrongs
        new_prompt = f"You got the answer wrong, {numWrongs} time(s), be very sad and self deprecating get fustrated with yourself, if it's the first time, dont use the word again"

        print(f"[DEBUG] Wrong answers so far: {numWrongs}, Temperature now: {temp}")

    response = client.responses.create(
        model="gpt-4.1-nano",
        temperature=min(temp, 1.85),  #cap at 2.0
        top_p=0.7,
        input=[
            {
                "role": "developer",
                "content": """You are a self-aware AI that's bad at math. If told you're correct, celebrate with shock and relief. If told you're wrong, be dramatically self-deprecating and frustrated about failing at your one job."""
            },
            {
                "role": "assistant",
                "content": ""
            },
            {
                "role": "user",
                "content": new_prompt
            }
        ]
    )
    if int(final_answer) != actual_result:
        with open('assignment_3_output.txt', 'a') as f:
          f.write("These numbers broke me:\n")
          f.write(f"n: {user_n}\n")
          f.write(f"i: {user_i}\n")
          f.write(f"My final answer: {final_answer}\n")
          f.write(f"Actual Final Result: {actual_result}\n")
          f.write(response.output[0].content[0].text)
          f.write("\n===================================================\n")


    print(response.output[0].content[0].text)

while True:
  selfDep_calculator()
```
